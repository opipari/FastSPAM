```python
class PointsRenderLayer(nn.Module):
    def __init__(
        self,
        raster_settings: Optional[PointsRasterizationSettings] = PointsRasterizationSettings(),
        compositor: Optional[nn.Module] = AlphaCompositor()
    ) -> None:
        super().__init__()
        
        rasterizer = PointsRasterizer(
            cameras=PerspectiveCameras(), 
            raster_settings=raster_settings
        )
        
        self.renderer = PointsRenderer(
            rasterizer=rasterizer,
            compositor=compositor
        )
        
    def set_cameras(
        self, 
        c: Type[CamerasBase]
    ) -> None:
        self.renderer.rasterizer.cameras = c
    
    def set_raster_settings(
        self,
        r: PointsRasterizationSettings
    ) -> None:
        self.renderer.rasterizer.raster_settings = r
    
    def set_raster_image_size(
        self,
        s: Union[int, Tuple[int, int]]
    ) -> None:
        self.renderer.rasterizer.raster_settings.image_size = s
        
    def forward(
        self,
        x: Pointclouds
    ) -> torch.Tensor:
        return self.renderer(x)
```

```python

def get_cameras(
    intrinsics: torch.Tensor,
    extrinsics: torch.Tensor,
    image_sizes: torch.Tensor
) -> Type[CamerasBase]:
    assert intrinsics.shape[0]==extrinsics.shape[0]==image_sizes.shape[0]
    assert intrinsics.shape[1:]==(4,4)
    assert extrinsics.shape[1:]==(4,4)
    assert image_sizes.shape[1:]==(2,)
    
    return PerspectiveCameras(in_ndc=False, K=intrinsics, R=extrinsics[:,:3,:3], T=extrinsics[:,:3,3], image_size=image_sizes)
    
class VideoPanopticMemoryCloud(nn.Module):
    def __init__(
        self,
        encoder: Optional[nn.Module] = None,
        decoder: Optional[nn.Module] = None,
        memory_window: Optional[int] = 1,
    ) -> None:
        super().__init__()
        
        # self.encoder = encoder
        # self.decoder = decoder
        self.memory_window = memory_window
        
        self.memory = PanopticMemoryCloud()
        self.renderer = PointsRenderLayer()

    def get_xy_depth(self,image_size, depth_map):
        assert depth_map.shape[2:]==image_size
        batch, _, height, width = depth_map.shape
        grid_h, grid_w = torch.meshgrid(torch.arange(height), torch.arange(width), indexing='ij')
        
        xy_map = torch.stack((grid_w, grid_h), dim=-1)
        print(xy_map.shape)
        
        xy_map = torch.tile(xy_map, (batch,1,1,1)).to(depth_map.device)
        xy_depth = torch.cat([xy_map, depth_map.permute(0,2,3,1)], axis=3)
        
        print(xy_depth.shape)
        
        
        return xy_depth
    
    def forward(
        self,
        image: torch.Tensor, # B x 3 x H x W
        depth: torch.Tensor, # B x 1 x H x W
        camera: Type[CamerasBase]
    ) -> torch.Tensor:
        B, _, H, W = image.shape
        image_size = (camera.image_size[0,0].item(), camera.image_size[0,1].item())
        assert (H,W)==image_size
        
        self.renderer.set_raster_image_size(image_size)
        self.renderer.set_cameras(camera)

        
        xy_depth = self.get_xy_depth(image_size, depth).reshape(B,-1,3)
        unprojected_points = camera.unproject_points(xy_depth)
        features = image.permute(0,2,3,1).reshape(B,-1,3)
        rendered_image = self.renderer(Pointclouds(points=unprojected_points, features=features).to('cuda'))
        fig, ax=plt.subplots(len(rendered_image),2)
        for i in range(len(rendered_image)):
            ax[i,0].imshow(image[i].permute(1,2,0).cpu()/255)
            ax[i,1].imshow(rendered_image[i].cpu()/255)
        plt.show()
        raise
        # image = self.render
        # current_features = self.encoder(visual_x)

        # get_pointcloud(current_memory)
        # render_pointcloud()
        # update_pointcloud()
        
        # memory_features = self.memory_cloud(pose_x)

        # out = self.decoder(torch.cat([current_features, memory_features]))
        # return out
```
